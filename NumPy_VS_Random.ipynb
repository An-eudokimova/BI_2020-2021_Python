```python
import numpy as np
import timeit
import random
import matplotlib.pyplot as plt
import seaborn as sns


time_np = [] 
time_random = []
n = 100000
xs = list(map(int, range(n)))

#numpy, defining function (for the next step)
def nprandom(x):
    return np.random.random(x)

#numpy, counting time for different n
for x in range(1, n + 1):
    time = timeit.timeit("nprandom(x)", setup="from __main__ import nprandom, x", number=1)
    time_np.append(time)

#random, defining function (for the next step)
def randomrandom(x):
    return random.random()

#random, counting time for different n
for x in range(1, n + 1):
    time = timeit.timeit("randomrandom(x)", setup="from __main__ import randomrandom, x", number=1)
    time_random.append(time)

#plotting
plt.rcParams['figure.figsize'] = [10, 6] #the size of image
plt.rcParams['font.size'] = 12 #the size of values
plt.rcParams['savefig.bbox'] = 'tight'

#style, color of plots
plt.style.use('seaborn-darkgrid')
col = plt.get_cmap('Set1')

plt.plot(xs, time_np, label = 'NumPy', color = col(1))
plt.plot(xs, time_random, label = 'Random', color = col(4))

plt.legend(loc = 2, fontsize = 17)
plt.title('NumPy VS Random: comparison of running time \n on the \
 example of generating random numbers of uniform distribution')
plt.ylabel('time, sec')
plt.xlabel('Number of random numbers')

```
