```python
import matplotlib.pyplot as plt

xs = list(map(int, range(21))) #list of independent variable
#I desided to do plot with three defferent line 
def fun1(x): #function1 for the first set of the dependent variabe
    return(2 * x + 5)
def fun2(x): #function2 for the second set of the dependent variabe
    return(x * (-1) + 2)
ys1 = list(map(fun1, xs)) #first set of the dependent variabe
ys2 = list(map(fun2, xs)) #second set of the dependent variabe
ys3 = list(map(int, range(21))) #third set of the dependent variabe

plt.rcParams['figure.figsize'] = [10, 6] #the size of image
plt.rcParams['font.size'] = 12 #the size of values
plt.rcParams['savefig.bbox'] = 'tight'

col = plt.get_cmap('Dark2') #set a color palette

plt.style.use('dark_background') #style

#plot lines
plt.plot(xs, ys1, linestyle='--', color = col(1), label = 'y = 2x + 5', linewidth=4)
plt.plot(xs, ys2, linestyle=':', color = col(0), label = 'y = -x + 2', linewidth=4)
plt.plot(xs, ys3, linestyle='-', color = col(2), label = 'y = x', linewidth=3)

plt.legend(loc = 2, fontsize = 17)
plt.title('Just some different lines', size=22)
plt.ylabel('y')
plt.xlabel('x')

plt.savefig('JustSomeLines.png', format='png')
```
