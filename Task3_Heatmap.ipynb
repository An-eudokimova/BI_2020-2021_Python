```python
#Here I will draw a graph on the real data that I am working with now. 
#It illustrates the coverage of contigs in various samples.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [10, 6] #the size of image
plt.rcParams['font.size'] = 12 #the size of values
plt.rcParams['savefig.bbox'] = 'tight'

#import dataset
df = pd.read_csv('coverage_msc31.csv', delimiter=',', index_col=0)
df = df.transpose()

sns.set_style("dark") #style

#heatmap
plot = sns.heatmap(df, cmap="Reds")
plt.title('Moscow centenarians', size = 22)
plt.xlabel('samples', size = 17)
plt.ylabel('contigs', size = 17)
plt.xticks(rotation=45)

plt.savefig('Task3_Heatmap.png', format='png')
```
